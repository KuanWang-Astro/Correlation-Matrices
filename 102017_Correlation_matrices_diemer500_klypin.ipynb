```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
from correlation_matrix import correlation
```

```python
diemer500_20p0_w = np.load('diemer500_klypin20p0_w.npz')
diemer500_20p0_wo = np.load('diemer500_klypin20p0_wo.npz')
diemer500_20p5_w = np.load('diemer500_klypin20p5_w.npz')
diemer500_20p5_wo = np.load('diemer500_klypin20p5_wo.npz')
diemer500_21p0_w = np.load('diemer500_klypin21p0_w.npz')
diemer500_21p0_wo = np.load('diemer500_klypin21p0_wo.npz')
```

```python
title_list = ['diemer500,-20.0','diemer500,-20.5','diemer500,-21.0']
```

```python
c = []

c.append(correlation(diemer500_20p0_w,diemer500_20p0_wo))
c.append(correlation(diemer500_20p5_w,diemer500_20p5_wo))
c.append(correlation(diemer500_21p0_w,diemer500_21p0_wo))
```

```python
fig = plt.figure(figsize=(30,10))
for i in range(1,4):
    fig.add_subplot(1,3,i)
    plt.title(title_list[i-1],fontsize=30)
    c[i-1].plot_correlation_matrix()
```
