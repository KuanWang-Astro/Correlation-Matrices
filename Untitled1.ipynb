```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def comb(mw,mwo):
    m = np.copy(mw)
    dim = m.shape[0]
    for i in range(dim):
        for j in range(dim):
            if i>j:
                m[i,j] = mwo[i,j]
    return m
```

```python
all_wp_wpcov = np.load('112017_wp_corr_yao&halotools.npz')
```

```python
def corr(cov):
    sigma = np.sqrt(cov.diagonal())
    return (cov/sigma).T/sigma
```

```python
title_list = ['chinchilla2560,-20.0','chinchilla2560,-20.5','chinchilla2560,-21.0',\
              'bolshoi,-20.0','bolshoi,-20.5','bolshoi,-21.0',\
              'chinchilla2048,-20.0','chinchilla2048,-20.5','chinchilla2048,-21.0',\
              'bolplanck,-20.0','bolplanck,-20.5','bolplanck,-21.0',\
              'chinchilla1024,-20.0','chinchilla1024,-20.5','chinchilla1024,-21.0',\
              'diemer250,-20.0','diemer250,-20.5','diemer250,-21.0',\
              'consuelo,-20.0','consuelo,-20.5','consuelo,-21.0',\
              'diemer500,-20.0','diemer500,-20.5','diemer500,-21.0',\
              'chinchilla512,-20.0','chinchilla512,-20.5','chinchilla512,-21.0']
```

```python
###Yao's code, fewer bins

fig = plt.figure(figsize=(30,90))
for i in range(1,28):
    fig.add_subplot(9,3,i)
    plt.title(title_list[i-1],fontsize=30)
    plt.imshow(comb(corr(all_wp_wpcov['wpcov_w_yao'][i-1]),corr(all_wp_wpcov['wpcov_wo_yao'][i-1])),cmap='Reds',vmax=1,vmin=0,interpolation='None')
    plt.colorbar()
    plt.axhline(5,0,1,c='k')
    plt.axvline(5,0,1,c='k')
    plt.axhline(6.5,0,1,c='b')
    plt.axvline(6.5,0,1,c='b')
```
