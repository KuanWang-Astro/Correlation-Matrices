```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
consuelo20_w = np.load('../Compute statistics new/101917_vpf_consuelo_w.npz')
consuelo20_wo = np.load('../Compute statistics new/101917_vpf_consuelo_wo.npz')
```

```python
consuelo20_w.files
```

```python
vpf_samep_w = []

for i in range(200):
    vpf_samep_w.append(consuelo20_w['vpf'][i::200])
```

```python
vpf_samep_wo = []

for i in range(200):
    vpf_samep_wo.append(consuelo20_wo['vpf'][i::200])
```

```python
vpf_corr_w = np.zeros((200,20,20))
vpf_corr_wo = np.zeros((200,20,20))

for i in range(200):
    vpf_corr_w[i] = np.corrcoef(vpf_samep_w[i].T)
    vpf_corr_wo[i] = np.corrcoef(vpf_samep_wo[i].T)
```

```python
def comb(matrix1,matrix2):
    matrix = np.copy(matrix1)
    dim = matrix.shape[0]
    for i in range(dim):
        for j in range(dim):
            if i>j:   ##lower left
                matrix[i,j] = matrix2[i,j]
    return matrix
```

```python
plt.imshow(comb(np.mean(vpf_corr_w,axis=0),np.mean(vpf_corr_wo,axis=0)),cmap='bwr',vmax=1,vmin=-1,interpolation='None')
plt.colorbar()
```

```python
plt.imshow(comb(np.mean(vpf_corr_w[:10],axis=0),np.mean(vpf_corr_wo[:10],axis=0)),cmap='bwr',vmax=1,vmin=-1,interpolation='None')
plt.colorbar()
```

```python
plt.imshow(comb(vpf_corr_w[0],vpf_corr_wo[0]),cmap='bwr',vmax=1,vmin=-1,interpolation='None')
plt.colorbar()
```
